Данные тестирования

Формат данных -- список кортежей, в кортеже -- число пакетов, время предобработки, время сравнения
Длина тела -- 100

Ключ -- все константные поля, для большого числа пакетов уместно. На 100000 пакетов групп около 5000.

```python
standard_single_generated: list[tuple[int, float, float]] = [
    (10000, 2.591230, 0.154435),
    (50000, 13.529948, 2.843051),
    (100000, 27.883509, 8.266063),
    (200000, 64.194047, 20.227330)
]
```

```python
standard_parallel_generated: list[tuple[int, float, float]] = [
    (10000, 5.381907, 1.916463), 
    (50000, 14.274566, 2.312220),
    (100000, 25.417172, 3.043305),
    (200000, 47.359136, 4.559206)
]
```

```python
tokenized_single_generated: list[tuple[int, float, float]] = [
    (10000, 2.665446, 0.152628), 
    (50000, 13.873506, 2.843259),
    (100000, 28.846628, 8.662931),
    (200000, 63.745986, 19.415005)
]
```

```python
tokenized_parallel_generated: list[tuple[int, float, float]] = [
    (10000, 5.784715, 1.861675), 
    (50000, 14.028417, 2.425722),
    (100000, 25.821579, 3.021785),
    (200000, 48.475127, 4.495626)
]
```

```python
import matplotlib.pyplot as plt
```

Зависимость времени предобработки от числа пакетов


```python
standard_single_x = [item[0] for item in standard_single_generated]
standard_single_y = [item[1] for item in standard_single_generated]

standard_parallel_x = [item[0] for item in standard_parallel_generated]
standard_parallel_y = [item[1] for item in standard_parallel_generated]

tokenized_single_x = [item[0] for item in standard_single_generated]
tokenized_single_y = [item[1] for item in tokenized_single_generated]

tokenized_parallel_x = [item[0] for item in tokenized_parallel_generated]
tokenized_parallel_y = [item[1] for item in tokenized_parallel_generated]

plt.plot(standard_single_x, standard_single_y, label="Посимвольное сравнение, последовательно", color="r", marker="o")
plt.plot(standard_parallel_x, standard_parallel_y, label="Посимвольное сравнение, параллельно", color="b", marker="o")
plt.plot(tokenized_single_x, tokenized_single_y, label="Токены, последовательно", color="g", marker="o")
plt.plot(tokenized_parallel_x, tokenized_parallel_y, label="Токены, параллельно", color="orange", marker="o")

plt.title("Время предобработки пакетов")
plt.xlabel("Число пакетов")
plt.ylabel("Время предобработки")

plt.legend()
plt.grid(True)
plt.show()
```

Зависимость времени постобработки от числа пакетов

```python
standard_single_x = [item[0] for item in standard_single_generated]
standard_single_y = [item[2] for item in standard_single_generated]

standard_parallel_x = [item[0] for item in standard_parallel_generated]
standard_parallel_y = [item[2] for item in standard_parallel_generated]

tokenized_single_x = [item[0] for item in tokenized_single_generated]
tokenized_single_y = [item[2] for item in tokenized_single_generated]

tokenized_parallel_x = [item[0] for item in tokenized_parallel_generated]
tokenized_parallel_y = [item[2] for item in tokenized_parallel_generated]

plt.plot(standard_single_x, standard_single_y, label="Посимвольное сравнение, последовательно", color="r", marker="o")
plt.plot(standard_parallel_x, standard_parallel_y, label="Посимвольное сравнение, параллельно", color="b", marker="o")
plt.plot(tokenized_single_x, tokenized_single_y, label="Токены, последовательно", color="g", marker="o")
plt.plot(tokenized_parallel_x, tokenized_parallel_y, label="Токены, параллельно", color="orange", marker="o")

plt.title("Время сравнения групп")
plt.xlabel("Число пакетов")
plt.ylabel("Время сравнения групп")

plt.legend()
plt.grid(True)
plt.show()
```

Данные для стандартных версий на 50000 пакетов, зависимость времени от длины пакета. В начале кортежа -- длина пакета, число символов в теле
